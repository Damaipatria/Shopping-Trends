```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Distribusi data berdasarkan kolom 'Gender'
data = df['Gender'].value_counts().rename_axis('Gender').reset_index(name='Count')

fig=px.pie(data, values='Count', names='Gender', title='Distribusi Gender', height=400)
fig.show()
```

```python
# 10 'Item Purchased' (Barang) paling populer
data = df['Item Purchased'].value_counts().rename_axis('Item Purchased').reset_index(name='Count')

fig = px.bar(data.head(10), x='Item Purchased', y='Count', color='Count', text_auto=True)
fig.update_traces(textposition="outside")
fig.show()
```

```python
# 10 'Item Purchased' (Barang) dengan Omzet paling banyak
data = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index(name='Purchase Amount (USD)').sort_values('Purchase Amount (USD)', ascending=False)

fig = px.bar(data.head(10), x='Item Purchased', y='Purchase Amount (USD)', color='Purchase Amount (USD)', text_auto='.3s')
fig.update_traces(textposition="outside")
fig.show()
```

```python
# 10 'Item Purchased' (Barang) paling populer pada Gender Male
data = df[df["Gender"].isin(["Male"])]['Item Purchased'].value_counts().rename_axis('Item Purchased').reset_index(name='Count')

fig = px.bar(data.head(10), x='Item Purchased', y='Count', color='Count', text_auto=True)
fig.update_traces(textposition="outside")
fig.show()
```

```python
# 10 'Item Purchased' (Barang) paling populer pada Gender Female
data = df[df["Gender"].isin(["Female"])]['Item Purchased'].value_counts().rename_axis('Item Purchased').reset_index(name='Count')

fig = px.bar(data.head(10), x='Item Purchased', y='Count', color='Count', text_auto=True)
fig.update_traces(textposition="outside")
fig.show()
```

```python
# Kategori paling populer
data = df['Category'].value_counts().rename_axis('Category').reset_index(name='Count')

fig = px.bar(data, x='Category', y='Count', color='Count', text_auto=True)
fig.update_traces(textposition="outside")
fig.show()
```

```python
# Ukuran paling populer
data = df['Size'].value_counts().rename_axis('Size').reset_index(name='Count')

fig = px.bar(data, x='Size', y='Count', color='Count', text_auto=True)
fig.update_traces(textposition="outside")
fig.show()
```

```python
# Perbandingan metode pembayaran yang disediakan dengan yang diinginkan customer
data1 = df['Payment Method'].value_counts().rename_axis('Payment Method').reset_index(name='Count')
data2 = df['Preferred Payment Method'].value_counts().rename_axis('Preferred Payment Method').reset_index(name='Count')

fig = go.Figure(data=[
    go.Bar(name='Payment Method', x=data1['Payment Method'], y=data1['Count'], marker_color='#b83289', text=data1['Count'], textposition='outside'),
    go.Bar(name='Preferred Payment Method', x=data2['Preferred Payment Method'], y=data2['Count'], marker_color='#4e03a1', text=data2['Count'], textposition='outside')
])
fig.update_layout(barmode='group')
fig.show()
```

```python
# 10 warna populer
data = df['Color'].value_counts().rename_axis('Color').reset_index(name='Count')

fig = px.bar(data.head(10), x='Color', y='Count', color='Count', text_auto=True)
fig.update_traces(textposition="outside")
fig.show()
```

```python
# perbandingan jumlah customer yang menggunakan diskon dengan yang tidak
data = df['Discount Applied'].value_counts().rename_axis('Discount').reset_index(name='Count')

fig=px.pie(data, values='Count', names='Discount', title='Perbandingan penggunaan diskon dengan tidak', height=400)
fig.show()
```
